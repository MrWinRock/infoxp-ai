## Object to array

```python
import re
import numpy as np

def parse_list(s: str) -> list[str]:
    if pd.isna(s) or s is None:
        return []
    return [p.strip() for p in str(s).split(',') if p.strip()]

def parse_developers(s: str) -> list[str]:
    if pd.isna(s) or not str(s).strip():
        return []
    s = str(s).strip()

    company_suffix = r'(Inc\.?|LLC\.?|Ltd\.?|Corporation|Corp\.?|Co\.?|Limited|AG|GmbH|S\.A\.?|PLC)'
    company_word   = r'(Games?|Entertainment|Interactive|Studios?|Software|Digital|Media|Productions?)'

    # If it matches a single-company pattern, keep whole string
    if re.search(rf'.+?\s+{company_suffix}$', s, re.I) or re.search(rf'.+?\s+{company_word}$', s, re.I):
        return [s]

    # Quoted segments → treat each quoted block as one entry
    if '"' in s:
        parts = [p.strip(' ,') for p in s.split('"') if p.strip(' ,')]
        return parts

    known_single = [
        'FromSoftware, Inc.', 'CAPCOM Co., Ltd.', 'KRAFTON, Inc.',
        'Bandai Namco Entertainment', 'Gaggle Studios, Inc.'
    ]
    if any(k in s for k in known_single):
        return [s]

    # If contains legal suffixes, split carefully and rejoin suffix tails
    if any(x in s for x in ['Inc.', 'Ltd.', 'Corp.', 'LLC', 'Corporation', 'Co.']):
        parts = [p.strip() for p in s.split(',')]
        devs, cur = [], ''
        for i, p in enumerate(parts):
            cur = f"{cur}, {p}"[2:] if cur else p
            if re.search(rf'\b{company_suffix}$', p, re.I) or i == len(parts) - 1:
                devs.append(cur)
                cur = ''
        return [d for d in devs if d]

    # Default: simple comma split
    return parse_list(s)

# Apply
df['Developers'] = df['Developers'].apply(parse_developers)
df['Categories'] = df['Categories'].apply(parse_list)
df['Genres']     = df['Genres'].apply(parse_list)
```
